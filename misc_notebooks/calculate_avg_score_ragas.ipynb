```python
import json
import math
```

```python
def calculate_average_faithfulness_score(json_path):
    with open(json_path, 'r', encoding='utf-8') as f:
        try:
            data = json.load(f)
        except json.JSONDecodeError:
            f.seek(0)
            data = [json.loads(line) for line in f]

    scores = []
    for entry in data:
        score = entry.get('faithfulness_score', None)
        # Check if score is a valid number and not -1.0 or nan
        if isinstance(score, (int, float)) and score != -1.0 and not math.isnan(score):
            scores.append(score)

    print("Scores used for average (nan and -1.0 removed):", scores)

    if not scores:
        print("No valid scores to average.")
        return None

    avg = sum(scores) / len(scores)
    print(f"Average faithfulness_score: {avg}")
    return avg
```

```python
calculate_average_faithfulness_score('../experiments/output/ragas/gpt4o_non_cot/hover_train_gpt4o_non_cot.results.json')
```
